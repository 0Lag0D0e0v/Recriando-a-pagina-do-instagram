```python
import pandas as pd
```

```python
produtos = pd.read_csv('https://docs.google.com/spreadsheets/d/1cMBZ7t2t7NTDa-oQoWuih1T9R7ZDSvrtV6I-ygtJ8Ls/export?format=csv')
```

```python
produtos.head()
```

```python
produtos.columns
```

```python
produtos['categoria'] = produtos['categoria'].str.replace('Sapato', 'Calçados')
```

```python
produtos.head()
```

```python
produtos[['tipo_produto', 'cor']] = produtos['nome_produto'].str.split(' ', n=1, expand=True)
```

```python
substituicoes = {
    'Fone': 'Fone de Ouvido',
    'Garrafa': 'Garrafa Térmica'
}

produtos['nome_produto'] = produtos['nome_produto'].replace(substituicoes, regex=True)
```

```python
produtos.head()
```

```python
produtos[produtos['nome_produto'].str.contains('Fone', case=False)]
```

```python
def corrigir_categoria(row):
    nome = row['nome_produto'].lower()

    if any(p in nome for p in ['fone', 'óculos']):
        return 'Eletrônicos'
    elif any(p in nome for p in ['mochila', 'bolsa']):
        return 'Acessórios'
    elif any(p in nome for p in ['tênis', 'camiseta', 'jaqueta', 'blusa', 'sandália', 'boné', 'meia']):
        return 'Vestuário'
    elif 'garrafa' in nome:
        return 'Utilidades Domésticas'
    else:
        return row['categoria']  # mantém a categoria atual se não reconhecer
produtos['categoria'] = produtos.apply(corrigir_categoria, axis=1)
```

```python
produtos.head()
```

```python
produtos['preco_unitario'] = (
    produtos['preco_unitario']
    .str.replace('R$', '', regex=False)
    .str.replace(',', '.', regex=False)
    .astype(float)
)
```

```python
def filtrar_por_preco(produtos, valor_minimo):
    return produtos[produtos['preco_unitario'] > valor_minimo]
filtrados = filtrar_por_preco(produtos, 200)
display(filtrados)
```

```python
def ordenar_por_coluna(produtos, coluna='preco_unitario', crescente=False):
    return produtos.sort_values(by=coluna, ascending=crescente)
ordenados = ordenar_por_coluna(produtos, 'preco_unitario', crescente=False)
display(ordenados)
```

```python
def resumo_por_categoria(produtos):
    return produtos.groupby('categoria')['preco_unitario'].sum().reset_index()
resumo = resumo_por_categoria(produtos)
display(resumo)
```
